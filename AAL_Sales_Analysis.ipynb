## 2. Data Analysis

Identifying higest and lowest revenue generating states

```python
data.groupby('State')['Sales'].sum().sort_values(ascending=False)
```

```python
data.groupby('Group')['Sales'].sum().sort_values(ascending=False)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Week']=data['Date'].dt.isocalendar().week
```

```python
data['Month']=data['Date'].dt.month
```

```python
weekly_sales =data.groupby('Week')['Sales'].sum().sort_values(ascending=False)
```

```python
monthly_sales =data.groupby('Month')['Sales'].sum().sort_values(ascending=False)
```
