# Lecture Notes on Basics of Matplotlib

Matplotlib is a widely used plotting library for Python that provides a flexible way to create various kinds of static, animated, and interactive plots.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Loading a Dataset

You can use pandas to load datasets from various formats. For example, loading a CSV file:

https://github.com/mwaskom/seaborn-data/blob/master/tips.csv

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head()
```

For this example, we will use the same built-in `tips` dataset.

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Basic Plot Types

### 1. Scatter Plot
A scatter plot displays values for two variables for a set of data points.

```python
plt.scatter(tips['total_bill'],tips['tip'])
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### 2. Line Plot
Line plots are used to represent data points in time or a continuous variable.

```python
plt.plot(tips['total_bill'],tips['tip'])
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### 3. Bar Plot
Bar plots display categorical data with rectangular bars.

```python
tips.groupby('day')['total_bill'].mean().plot(kind='bar')
```

### 4. Histogram
Histograms display the distribution of a single variable.

### 5. Box Plot
Box plots show the distribution of quantitative data and can be used to compare between categories.

### 6. Heatmap (Using Matplotlib with Seaborn)
Heatmaps display data in matrix form using color coding.

### 7. Pair Plot (Using Matplotlib with Seaborn)
Pair plots are used to plot pairwise relationships in a dataset.
